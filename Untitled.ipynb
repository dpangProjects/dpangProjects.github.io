```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import time
```

```python
columns=['Name','Codename','Cores','Clock','Socket','Process','L3 Cache','TDP','Released']
df=pd.DataFrame()
r=requests.get('https://www.techpowerup.com/cpu-specs/?released=2000&sort=released')
root = BeautifulSoup(r.text, 'html')
table = root.find("table")
other=pd.read_html(str(table))[0]
df=other
df = df.iloc[: , :-1]
df = df.set_axis(columns, axis=1)
```

```python
url='https://www.techpowerup.com/cpu-specs/?released='
rest='&sort=released'
years=[]
for i in range(2001,2023):
    print(i)
    r=requests.get(url+str(i)+rest)
    root = BeautifulSoup(r.text, 'html')
    table = root.find("table")
    other = pd.read_html(str(table))[0]
    other = other.iloc[: , :-1]
    other = other.set_axis(columns, axis=1)
    df=pd.concat([df,other])
    time.sleep(3)
```

```python
columns=['Product Name','GPU Chip','Release Date','Bus','Memory','GPU Clock','Memory Clock','Shaders/TMUs/ROPs']
df=pd.DataFrame()
r=requests.get('https://www.techpowerup.com/gpu-specs/?released=2000&sort=released')
root = BeautifulSoup(r.text, 'html')
table = root.find_all("table")
other=pd.read_html(str(table[1]))[0]
df=other
df = df.set_axis(columns, axis=1)
```

```python
url='https://www.techpowerup.com/gpu-specs/?released='
rest='&sort=released'
years=[]
for i in range(2001,2023):
    print(i)
    r=requests.get(url+str(i)+rest)
    root = BeautifulSoup(r.text, 'html')
    table = root.find_all("table")
    other=pd.read_html(str(table[1]))[0]
    other = other.set_axis(columns, axis=1)
    df=pd.concat([df,other])
    time.sleep(3)
```

```python
compression_opts = dict(method='zip',
                        archive_name='CPUs.csv')  
df.to_csv('CPUs.zip', index=False,
          compression=compression_opts)  
compression_opts = dict(method='zip',
                        archive_name='GPUs.csv')  
df.to_csv('GPUs.zip', index=False,
          compression=compression_opts)  
```

```python
cpu_df=pd.read_csv("CPUs.csv")
gpu_df=pd.read_csv("GPUs.csv")
```

```python
from datetime import date
from datetime import datetime
```

```python
cpu_df = cpu_df[cpu_df['Released'] != 'Never Released']
cpu_df = cpu_df[cpu_df['Released'] != 'Unknown']
dates=[]
for i,row in cpu_df.iterrows():
    row['Released'] = row['Released'].replace('st', '').replace('rd', '').replace('nd', '').replace('th', '')
    if len(row['Released'])==8:
        dt_obj = datetime.strptime(row['Released'], '%b %Y')
    elif len(row['Released'])==4:
        dt_obj = datetime.strptime(row['Released'], '%Y')
    else:
        dt_obj = datetime.strptime(row['Released'], '%b %d, %Y')
    dates.append(dt_obj.date())
```

```python
cpu_df.plot.scatter(x='datetime_Released',y='Actual Cores')
```

```python
cpu_df['Actual Cores']=cpu_df['Actual Cores'].astype(float)
```

```python
import re
cores=[]
for i,row in cpu_df.iterrows():
    if len(row['Cores'])==1 or len(row['Cores'])==2:
        result = re.search(r"(\d+)", row['Cores'])
    else:
        result = re.search(r"(\d+) \/ (\d+)|(\d+)", row['Cores'])
    cores.append(result.group(1))
cpu_df['Actual Cores']=cores  
```
